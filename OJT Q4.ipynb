```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

# Polynomial Plotting with Annotations:

```python
np.random.seed(50)
coefficients = np.random.randint(-10, 11, size=6) 
def polynomial(x):
    return np.polyval(coefficients, x)

x_values = np.linspace(-5, 5, 100)

y_values = polynomial(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Polynomial')
plt.title('Plot of a Random 5th Degree Polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

# Annotate the plot to identify the maxima and minima of the polynomial.

```python
# Find peaks (maxima and minima)
peaks, _ = find_peaks(y_values)

# Plot the polynomial
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Polynomial')
plt.scatter(x_values[peaks], y_values[peaks], color='red', marker='o', label='Maxima/Minima')

# Annotate the maxima and minima
for peak in peaks:
    plt.annotate(f'({x_values[peak]:.2f}, {y_values[peak]:.2f})', (x_values[peak], y_values[peak]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Plot of a Polynomial with Annotated Maxima and Minima')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
